```python
import pandas as pd
import numpy as np
import requests
```

The data is pulled from an API link found on https://apipheny.io/free-api/. The data is in JSON format and is converted to a list.

```python
universityInfo = requests.get("http://universities.hipolabs.com/search?country=United+States")
universityData = universityInfo.json()
```

```python
print(universityData)
```

```python
print(type(universityInfo))
print(type(universityData))
```

The list contains a dictionary inside a list data set. The list is converted to a dataframe.

```python
university = pd.DataFrame(universityData)
print(type(university))
```

```python
university.head()
```

The first descriptive statistic, display the total null values for each column in the data set.

```python
university.isnull().sum()
```

The second descriptive statistic displays information about the data set including: the range index and total number of rows in the data set (4362 entries), total number of columns in the data set (6), the name of each column header, the non-null values in each column, and the data type of each column.

```python
university.info()
```

A data query selects and prints all universities in the list that are in the state of Florida. Note: Only 94 universities in this list had a valid state. The rest were "None" values.

```python
print(university.query("`state-province` == 'Florida'"))
```
